Összes adat beolvasása

```python
#összes megadott csv beolvasás
import pandas as pd
import glob
import numpy as np
import matplotlib.pyplot as plt


all_files = glob.glob('data/*.csv')

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

df = pd.concat(li, axis=0, ignore_index=True)

df['letoltes_datum'] = pd.to_datetime(df.letoltes_datum, format='%Y-%m-%d')
df['tweetCreated'] = pd.to_datetime(df.tweetCreated, format='%Y-%m-%d')
df['userCreateDt'] = pd.to_datetime(df.userCreateDt, format='%Y-%m-%d')
```

Lokációs adatok a csv-k ből

```python
location = df.groupby(['userLocation'])['tweetID'].count().reset_index(name ='db').sort_values(by='db', ascending=False)

print(location.head())
```

Lokációs adatok és a forgatótábla összekulcsolása

```python
lokacio = location.set_index('userLocation').join(man.set_index('userLocation')).sort_values(by='db', ascending=False).reset_index()

```
